```python
animals = ['Great White Shark', 'Blue Whale', 'African Elephant',
           'Bald Eagle', 'Orangutan', 'Tiger', 'Panda', 'Koala']

for animal in animals:
    print(animal)
```

```python
i = 0
while i < 5:
    print(i)
    i += 1
```

```python
for animal in animals:
    if animal == 'Orangutan':
        break
    print(animal)
```

```python
for animal in animals:
    if animal == 'Bald Eagle':
        continue
    print(animal)
```

```python
for animal in animals:
    if animal == 'Panda':
        pass
    else:
        print(animal)
```
